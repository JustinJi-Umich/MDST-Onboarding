## Feature Engineering

You may want to convert categorical variables to numerical. For example, education takes on the value Graduate and Not Graduate. But we want it to be 0 or 1 for machine learning algorithms to use.

```python
loan_data['education'] = loan_data['education'].map({'Graduate': 1, 'Not Graduate': 0})
# Hint: Other categorical variables are self_employed and loan_status
# TODO

# 1. Reload the data just to be safe (resets everything to original text)
loan_data = pd.read_csv("loan_approval.csv")

# 2. Convert 'education'
loan_data['education'] = loan_data['education'].map({'Graduate': 1, 'Not Graduate': 0})

# 3. Convert 'self_employed'
loan_data['self_employed'] = loan_data['self_employed'].map({'Yes': 1, 'No': 0})

# 4. Convert 'loan_status'
# We use .str.strip() to handle any hidden spaces in " Approved " or " Rejected "
loan_data['loan_status'] = loan_data['loan_status'].str.strip().map({'Approved': 1, 'Rejected': 0})

# 5. Check the result
loan_data.head()
```

## Model Selection

You are free to use any classification machine learning models you like: Logistic Regression, Decision Trees/Random Forests, Support Vector Machines, KNN ...

```python
# TODO


# Again same start with tutorial and checkpoint 1, we start examining the correlation between x and y.
# in this case x is the input, and y is the loan answer / status
X = loan_data.drop(columns=['loan_id', 'loan_status'])
y = loan_data['loan_status']

# and i am going to do 80% studying (training), and 20% testing this time
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=2)

# Initialize the Random Forest model
model = RandomForestClassifier(random_state=2)
```

## Model Training and Evaluation

```python
from sklearn.metrics import accuracy_score, classification_report
#imported some fancy stuff from the python machine learning library

# First I am going to fit it
model.fit(X_train, y_train)

#Using the predict function of google colab
y_pred = model.predict(X_test)

# return a number for th accuracy score
print("Accuracy Score:", accuracy_score(y_test, y_pred))
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## Model Optimization and Testing

```python


# lets say 3 settings for the "trees"
param_grid = {'n_estimators': [50, 100, 200]}

grid_search = GridSearchCV(RandomForestClassifier(random_state=42), param_grid, cv=3)
grid_search.fit(X_train, y_train)

# Print all the stuff out
print("Best Number of Trees:", grid_search.best_params_)
print("Best Accuracy:", grid_search.best_score_)

#And the accuracy is pretty good using those parameters, which is about 97.7%
```
